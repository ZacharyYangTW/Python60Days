```python
#1. 畫出 cos 圖檔，並儲存
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)

#plt.show()
plt.savefig("filename.png",dpi=300,format="png")
```

```python
#2. 給定散點圖顏色
import matplotlib.pyplot as plt
import numpy as np

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
T = np.arctan2(Y,X)
plt.scatter(X,Y, s=75, c=T, alpha=.5)
plt.grid(True)
plt.show()
#correct
```
