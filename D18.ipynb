```python
#1. 畫出 cos 圖檔，並儲存
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)

#plt.show()
plt.savefig("filename.png",dpi=300,format="png")
```
