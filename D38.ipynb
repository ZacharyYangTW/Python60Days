```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display

import pingouin as pg
import researchpy   
%matplotlib inline
```

```python
df_train = pd.read_csv("Titanic_train.csv")
print(df_train.info())
```

```python
## 這邊我們做一個調整，把 Survived 變成離散型變數 Survived_cate

Survived_cate=df_train.copy() #複製一份資料給 Survived_cate
#轉型 
#透過 preprocessing 轉換
from sklearn import preprocessing
#下面兩行程式，讓 Survived=1, Survived-0.
le = preprocessing.LabelEncoder()
Survived_cate['Survived']=le.fit_transform(Survived_cate['Survived'])
display(Survived_cate)
```

```python
## 這邊我們做一個調整，把 Sex 變成離散型變數 Sex

#轉型 
#下面兩行程式，讓 Male=1, FeMale-0.
le = preprocessing.LabelEncoder()
Survived_cate['Sex']=le.fit_transform(Survived_cate['Sex'])
display(Survived_cate)
```

```python
#Zach Age: 連續型
#Zach Survived_cate: 離散型
#ZACH Sex: 離散型
#Zach Fare: 連續型
#Q1: 連續age vs 離散Survived_cate
aov = pg.anova(dv='Age', between='Survived', data=Survived_cate, detailed=True)
aov
```

```python
#correct
#計算相關係數，不能允許有遺失值，所以必須先補值，或者把遺失值刪除
## 取出資料後，把遺失值刪除
complete_data=Survived_cate[['Age','Survived']].dropna()
display(complete_data)
```

```python
etaSq = aov.SS[0] / (aov.SS[0] + aov.SS[1])
etaSq
```

```python
def judgment_etaSq(etaSq):
    if etaSq < .01:
        qual = 'Negligible'
    elif etaSq < .06:
        qual = 'Small'
    elif etaSq < .14:
        qual = 'Medium'
    else:
        qual = 'Large'
    return(qual)
judgment_etaSq(etaSq)
```

```python
crosstab, res = researchpy.crosstab(Survived_cate['Sex'], Survived_cate['Survived'], test='chi-square')
#print(res)
print("Cramer's value is",res.loc[2,'results'])

#這邊用卡方檢定獨立性，所以採用的 test 參數為卡方 "test =" argument.
# 採用的變數在這個模組中，會自己根據資料集來判斷，Cramer's Phi if it a 2x2 table, or Cramer's V is larger than 2x2.
```

```python
## 寫一個副程式判斷相關性的強度
def judgment_CramerV(df,V):
    if df == 1:
        if V < 0.10:
            qual = 'negligible'
        elif V < 0.30:
            qual = 'small'
        elif V < 0.50:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 2:
        if V < 0.07:
            qual = 'negligible'
        elif V < 0.21:
            qual = 'small'
        elif V < 0.35:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 3:
        if V < 0.06:
            qual = 'negligible'
        elif V < 0.17:
            qual = 'small'
        elif V < 0.29:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 4:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.15:
            qual = 'small'
        elif V < 0.25:
            qual = 'medium'
        else:
            qual = 'large'
    else:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.13:
            qual = 'small'
        elif V < 0.22:
            qual = 'medium'
        else:
            qual = 'large'
    return(qual)

judgment_CramerV(df,res.loc[2,'results'])
```
